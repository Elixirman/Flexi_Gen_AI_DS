# About


Demonstration of code snippets for creating basic plots and visualizing datasets using Matplotlib and Seaborn.

This includes creating line plots, scatter plots, histograms, and box plots.



# Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Generation

```python
# Set seed for reproducibility
np.random.seed(42)

# Simulate data
time = np.arange(1, 101)  # Create an array of integers from 1 to 100 (inclusive) representing time points
value = np.sin(time) + np.random.normal(scale=0.5, size=100)  # Generate sine values with added random noise
x = np.random.normal(size=100)  # Generate 100 random values from a normal distribution for x-coordinates
y = np.random.normal(size=100)  # Generate 100 random values from a normal distribution for y-coordinates
category = np.random.choice(['A', 'B', 'C'], size=100)  # Randomly assign one of three categories (A, B, or C) to each observation

# Create DataFrame
df = pd.DataFrame({'time': time, 'value': value, 'x': x, 'y': y, 'category': category})  # Combine the arrays into a DataFrame

df.head()  # Display the first few rows of the DataFrame
```

# Matplotlib Plots

## Line Plot

```python
import matplotlib.pyplot as plt

# Line plot
plt.figure(figsize=(10, 5))  # Set the figure size
plt.plot(df['time'], df['value'], marker='o', color='g')  # Plot 'value' over 'time' with circle markers
plt.title('Line Plot of Value Over Time')  # Add a title to the plot
plt.xlabel('Time')  # Label the x-axis
plt.ylabel('Value')  # Label the y-axis
plt.ylabel('Value')  # Label the y-axis
plt.grid(True)  # Add a grid for better readability
plt.show()  # Display the plot
```

## Scatter Plot

```python
# Scatter plot
plt.figure(figsize=(10, 5))  # Set the figure size
plt.scatter(df['x'], df['y'], c='red', alpha=0.8)  # Create a scatter plot of 'x' vs 'y' with blue points and 50% transparency
plt.title('Scatter Plot of X vs Y')  # Add a title to the plot
plt.xlabel('X')  # Label the x-axis
plt.ylabel('Y')  # Label the y-axis
plt.grid(True)  # Add a grid for better readability
plt.show()  # Display the plot
```

# Seaborn Plot

## Histograms

```python
plt.figure(figsize=(10, 5))  # Set the figure size
sns.histplot(df['value'], kde=True)  # Create a histogram of 'value' with a kernel density estimate (KDE) line
plt.title('Histogram of Value')  # Add a title to the plot
plt.xlabel('Value')  # Label the x-axis
plt.ylabel('Frequency')  # Label the y-axis
plt.show()  # Display the plot
```

## Box Plots

```python
plt.figure(figsize=(10, 5))  # Set the figure size
sns.boxplot(x='category', y='value', data=df)  # Create a box plot of 'value' grouped by 'category'
plt.title('Box Plot of Value by Category')  # Add a title to the plot
plt.xlabel('Category')  # Label the x-axis
plt.ylabel('Value')  # Label the y-axis
plt.show()  # Display the plot
```
